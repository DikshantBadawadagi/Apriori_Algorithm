```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from scipy.cluster.hierarchy import dendrogram, linkage
from sklearn.preprocessing import StandardScaler
labels = ['A', 'B', 'C', 'D', 'E']
x_values = [2, 3, 1, 3, 1.5]
y_values = [2, 2, 1, 1, 0.5]
data = pd.DataFrame({'Label': labels, 'X': x_values, 'Y': y_values})
data_array = data[['X', 'Y']].values
scaler = StandardScaler()
data_scaled = scaler.fit_transform(data_array)
linked = linkage(data_scaled, method='ward')
plt.figure(figsize=(10, 5))
dendrogram(linked,
           orientation='top',
           labels=data['Label'].values,
           distance_sort='descending',
           show_leaf_counts=True)
plt.title('Hierarchical Clustering Dendrogram')
plt.xlabel('Objects')
plt.ylabel('Distance')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from scipy.cluster.hierarchy import dendrogram, linkage
from sklearn.preprocessing import StandardScaler

# Create a small dataset
labels = ['A', 'B', 'C', 'D', 'E']
x_values = [2, 3, 1, 3, 1.5]
y_values = [2, 2, 1, 1, 0.5]
data = pd.DataFrame({'Label': labels, 'X': x_values, 'Y': y_values})

# Prepare data for clustering
data_array = data[['X', 'Y']].values

# Scale the data
scaler = StandardScaler()
data_scaled = scaler.fit_transform(data_array)

# Perform hierarchical clustering using single linkage
linked = linkage(data_scaled, method='single')

# Create the dendrogram
plt.figure(figsize=(10, 5))
dendrogram(linked,
           orientation='top',
           labels=data['Label'].values,
           distance_sort='descending',
           show_leaf_counts=True)

plt.title('Single Linkage Hierarchical Clustering Dendrogram')
plt.xlabel('Objects')
plt.ylabel('Distance')
plt.show()
```
